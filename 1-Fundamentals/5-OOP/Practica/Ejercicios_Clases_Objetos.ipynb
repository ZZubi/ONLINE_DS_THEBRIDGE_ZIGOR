## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomésticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas últimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un número con todas las ventas.
* Método que calcula la media de ventas de los últimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
## 01
class Tienda:
    '''
    Documentación de la clase Tienda

    Parameters:
        nombre:str
        direccion:int
        num_empleados:int
        ventas_ultimos_3_meses:list
    '''
    tipo = "Electrodomésticos"
    abierta = True
    nombre:str = None
    direccion:int = None
    num_empleados:int = None 
    ventas_ultimos_3_meses:list = None

    def __init__(self, nombre:str, direccion:int, num_empleados:int, ventas_ultimos_3_meses:list):
        '''
        Documentacion del init

        Parameters:
            nombre:str
            direccion:int
            num_empleados:int
            ventas_ultimos_3_meses:list
        '''
        self.nombre = nombre
        self.direccion = direccion
        self.num_empleados = num_empleados
        self.ventas_ultimos_3_meses = ventas_ultimos_3_meses

    def get_suma_de_ventas_todos_los_meses(self):
        '''
        Documentacion de get_suma_de_ventas_todos_los_meses
        '''
        return sum(self.ventas_ultimos_3_meses)

    def get_media_de_ventas_por_empleado_ultimos_3_meses(self):
        '''
        Documentacion de get_media_de_ventas_por_empleado_ultimos_3_meses
        '''
        return sum(self.ventas_ultimos_3_meses) / len(self.ventas_ultimos_3_meses) / self.num_empleados

    def get_nombre_y_direccion(self):
        '''
        Documentacion de get_nombre_y_direccion
        '''
        return f"Nombre: '{self.nombre}' / Dirección: '{self.direccion}'"
    
    def get_ventas_del_ultimo_mes(self):
        '''
        Documentacion de get_ventas_del_ultimo_mes
        '''
        return self.ventas_ultimos_3_meses[-1]

    def get_and_set_proyeccion_de_ventas(self, dinero_invertido_en_marketing:float):
        '''
        Documentacion de get_and_set_proyeccion_de_ventas

        Paramenter:
            dinero_invertido_en_marketing:float
        '''
        if dinero_invertido_en_marketing < 1000:
            self.ventas_ultimos_3_meses = [ventas_mes*1.2 for ventas_mes in self.ventas_ultimos_3_meses]
        else:
            self.ventas_ultimos_3_meses = [ventas_mes*1.5 for ventas_mes in self.ventas_ultimos_3_meses]

        return self.ventas_ultimos_3_meses
    
## 02
tienda_1 = Tienda('Mi primera tienda', "Calle Mateo Moraza, 3 - Vitoria", 8, [35000, 40000, 38000])
tienda_2 = Tienda('Mi segunda tienda', "Avenida de los Huetos, 8 - Vitoria", 15, [31000, 11000, 3000])
tienda_3 = Tienda('Mi tercera tienda', "Calle Kalea, 3 - Vitoria", 4, [3500, 6000, 3500])

## 03
print("\r\n", "03_______________________________________________")
print(tienda_1.tipo)
print(tienda_1.abierta)
print(tienda_1.nombre)
print(tienda_1.direccion)
print(tienda_1.num_empleados)
print(tienda_1.ventas_ultimos_3_meses)
print(tienda_1.get_suma_de_ventas_todos_los_meses())
print(tienda_1.get_media_de_ventas_por_empleado_ultimos_3_meses())
print(tienda_1.get_nombre_y_direccion())
print(tienda_1.get_ventas_del_ultimo_mes())
print(tienda_1.get_and_set_proyeccion_de_ventas(900))
print(tienda_1.get_and_set_proyeccion_de_ventas(1100))

## 04
print("\r\n", "04_______________________________________________")

mis_tiendas = [tienda_1, tienda_2, tienda_3]
for tienda in mis_tiendas:
    print(tienda.get_ventas_del_ultimo_mes())

## 05
print("\r\n", "05_______________________________________________")
print("Tiendas que se encuentran en una avenida:")
for tienda in mis_tiendas:
    if "avenida" in tienda.direccion.lower():
        print(tienda.nombre)

## 06
print("\r\n", "06_______________________________________________")
print(Tienda.__doc__)
print(Tienda.__init__.__doc__)
print(Tienda.get_and_set_proyeccion_de_ventas.__doc__)
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
## 01
class Perro:
    '''
    Documentacion de la clase Perro

    Paramenters:
        patas:int
        orejas:int
        ojos:int 
        velocidad:int
        raza:str
        color_pelo:str
        dueno:str
    '''
    patas:int = 4
    orejas:int = 2
    ojos:int = 2 
    velocidad:int = 0

    def __init__(self, raza:str, color_pelo:str = "Marrón", dueno:str = None):
        '''
        Documentacion de __init__

        Paramenters:
            raza:str
            color_pelo:str -> Default "Marrón"
            dueno:str -> Default None
        '''
        self.raza = raza
        self.color_pelo = color_pelo
        self.dueno = dueno

    def andar(self, aumento_velocidad:int):
        '''
        Documentacion de andar()

        Paramenters:
            aumento_velocidad:int
        '''
        self.velocidad += aumento_velocidad

    def parar(self):
        '''
        Documentacion de parar()

        Paramenters:
            None
        '''
        self.velocidad = 0

    def ladrar(self, entrada:str):
        '''
        Documentacion de ladrar

        Paramenters:
            None
        '''
        return "GUAU!" + entrada


## 02
perro_sin_dueno = Perro("Pastor alemán")

## 03
print("\r\n", "03_______________________________________________")
print(perro_sin_dueno.patas)
print(perro_sin_dueno.orejas)
print(perro_sin_dueno.ojos)
print(perro_sin_dueno.velocidad)
print(perro_sin_dueno.raza)
print(perro_sin_dueno.color_pelo)
print(perro_sin_dueno.dueno)

## 04
print("\r\n", "04_______________________________________________")
print(perro_sin_dueno.velocidad)
perro_sin_dueno.andar(2)
print(perro_sin_dueno.velocidad)
perro_sin_dueno.andar(2)
print(perro_sin_dueno.velocidad)
perro_sin_dueno.andar(7)
print(perro_sin_dueno.velocidad)


## 05
print("\r\n", "05_______________________________________________")
print(perro_sin_dueno.velocidad)
perro_sin_dueno.parar()
print(perro_sin_dueno.velocidad)

print("\r\n", "06_______________________________________________")
print(Perro.__doc__)
print(Perro.__init__.__doc__)
print(Perro.andar.__doc__)
```
